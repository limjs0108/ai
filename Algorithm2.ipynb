# 두 변수의 데이터를 교환

```python
# 합계값을 계산하는 알고리즘

# 첨자 index   0    1   2    3    4
# 데이터       12  13  11  14  10

# 데이터들의 합계를 구하는 가장 단순한 알고리즘은 sum

# arr[0] + arr[1] + arr[2] + arr[3] + arr[4]

# 로도 표현할 수 있다. 그렇지만 배열의 개수가 늘어나면 사용하기 힘들다.
# 반복 구조로 변형하자.

# arr[0] -> sum
# sum + arr[1] -> sum
# sum + arr[2] -> sum
# sum + arr[3] -> sum
# sum + arr[4] -> sum

arr = [12, 13, 11, 14, 10]
sum = 0
for i in range(len(arr)):
    sum += arr[i]
print(sum)
```

```python
# 최대값을 찾는 알고리즘

arr = [12, 13, 11, 14, 10]
max = 0

for i in arr:
    if i > max:
        max = i
        
print(max)
```

```python
# Linear Search 선형 탐색법

# 보관된 데이터를 맨 앞부터 순서대로 원하는 값을 찾을 때까지 탐색한다.
# 탐색 처리는 반복 구조를 사용하요 기술한다.
# 반복 구조에서는 종료 조건을 잊지 말자.

# 첨자(index)   0  1  2  3  4
# 데이터         4  2  3  5  1     찾을 데이터 : 5

idx = [4, 2, 3, 5, 1]
data1 = int(input('찾으려는 데이터 입력 : '))
for i in range(len(idx)):
    if idx[i] == data1:
        print(f'{i}번째 요소가 일치')
        break
    else:
        print('찾지 못했습니다.')
    
```

```python
def linear_search(no, list):
    for i in range(len(list)):
        if no == list[i]:
            print(f'{i}번째 요소가 일치')
            return
    print('찾지 못했습니다.')
    return
    
print(linear_search(5, [4, 2, 3, 5, 1]))
```
