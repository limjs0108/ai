```python
import requests
```

```python
from bs4 import BeautifulSoup as bs
```

```python
import pandas
```

```python
html = requests.get('https://movie.naver.com/movie/sdb/rank/rmovie.naver')
```

```python
soup = bs(html.text)
```

```python
len(soup.select('tr > td.title'))
```

```python
movies = soup.select('tr > td.title')
movies[0]
```

```python
title = movies[0].select('div.tit3 > a')[0].text
title
```

```python
movie_data = []
rank = 1
for movie in movies:
  title = movie.select('div.tit3 > a')[0].text
  movie_data.append([rank, title])
  rank += 1

movie_data
```

```python
import pandas as pd

df = pd.DataFrame(movie_data, columns = ['순위', '타이틀'])
df
```

```python
df.to_excel('Naver_movies.xlsx',index=False)
```
