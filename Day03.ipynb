```python
score = input('당신의 점수를 입력하세요 ')

if int(score[-2]) > 8:
  grade = 'A'
elif int(score[-2]) > 7:
  grade = 'B'
elif int(score[-2]) > 6:
  grade = 'C'
elif int(score[-2]) > 5:
  grade = 'D'
else:
  grade = 'F'

if int(score[-1]) > 6:
  grade1 = '+'
elif int(score[-1]) > 3:
  grade1 = '0'
else:
  grade1 = '-'

print(f'당신의 점수는 {score} 점이고 학점은 {grade}{grade1}학점입니다.')
```

```python
# while 중첩
# while 내부에 또 다른 while 을 사용하는 것을 중첩이라고 한다.

d = 1
while d <= 5:
  c = 1
  while c <= 3:
    print(f'{d}일차 {c}교시입니다.')
    c += 1
  d += 1
```

```python
##########
# for
##########

# 몇 번 반복 시킬지 그 횟수가 명확할 때 주로 사용한다.

# for 변수 in 반복시킬값들:
#   반복실행문

# 리스트
# for i in [1,2,3]:
#   print(i)

# 문자열
# for i in 'hello':
#   print(i)

# 튜플
# for i in(1,2,3):
#   print(i)

# range()
for i in range(1,4,2):      # range(시작(생략 가능, 끝(미만),스텝(생략 가능))) 
  print(i)

# set, dict 는 순서와 상관 없이는 사용 가능
```

```python
pw = input('비밀 번호를 입력하세요 >>> ')
cnt_al = 0
cnt_num = 0

for i in pw:
  if i.isalpha():
    cnt_al += 1
  elif i.isnumeric():
    cnt_num += 1
 
if cnt_al == 0 or cnt_num == 0:
  print('불가능한 비번입니다.')
else:
  print('사용 가능한 비번입니다')   


```
