```python
import matplotlib.patches as patches # 사진 수정 관룐 모듈
import matplotlib.image as mpimg     # 사진 읽어오는 모듈 
import matplotlib.pyplot as plt      # 사진 관련 기본 모듈

img = mpimg.imread('img2.jpg')  # 사진 읽어오기
fig, ax = plt.subplots()        # 사진을 표시할 액자와 사진 설정
ax.imshow(img)                  # 액자에 사진 삽입
for i in range(len(parsed['faces'])):
  x,y,w,h = parsed['faces'][i]['roi'].values()  # 인식한 얼굴의 roi 위치정보를 변수로 할당
  age, age_conf = parsed['faces'][i]['age'].values()
  gender, gender_conf = parsed['faces'][i]['gender'].values()
  anno = gender + ', ' + age
  re_face = patches.Rectangle((x,y),w,h,facecolor = 'none',edgecolor ='r')  # 그 위치에 사각형을 표시
  plt.text(x,y+250, anno, color ='blue') # 글자 설정
  ax.add_patch(re_face)
```
