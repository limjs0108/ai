```python
!pip install selenium

from selenium import webdriver

from bs4 import BeautifulSoup as bs
!apt-get update
!apt install chromium-chromedriver
!cp /usr/lib/chromium-browser/chromedriver /usr/bin

options = webdriver.ChromeOptions()
options.add_argument('--headless')
options.add_argument('--no-sandbox')
options.add_argument('--disable-dev-shm-usage')
```

```python
driver1 = webdriver.Chrome('chromedriver',options=options)
driver1.get('https://www.genie.co.kr/chart/top200')
```

```python
html1 = driver1.page_source
soup1 = bs(html1)

```

```python
songs1 = []
songs1 = soup1.select('tr.list')



```

```python
songs_data =[]
rank = 1
songs = soup1.select('tr.list')
for song in songs:
    title = song.select('a.title')[0].text.strip()
    singer = song.select('a.artist')[0].text.strip()
    songs_data.append(['지니', rank, title, singer])
    rank += 1
    
songs_data
```

```python
driver2 = driver1 = webdriver.Chrome('chromedriver',options=options)
driver2.get('https://www.genie.co.kr/chart/top200?ditc=D&ymd=20220509&hh=10&rtm=Y&pg=2')
```

```python
html2 = driver2.page_source
soup2 = bs(html2)
```

```python
songs2 = []
songs2 = soup2.select('tr.list')
```

```python
rank = 51
songs1 = soup2.select('tr.list')
for song in songs1:
    title = song.select('a.title')[0].text.strip()
    singer = song.select('a.artist')[0].text.strip()
    songs_data.append(['지니', rank, title, singer])
    rank += 1
    
songs_data
```

```python
import pandas as pd
df = pd.DataFrame(songs_data,columns = ['차트명','순위','제목','가수'])
df
```
